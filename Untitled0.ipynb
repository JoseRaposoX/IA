```python
# ============================================
# PROJETO: ESPECULAÃ‡ÃƒO DE CRIPTOMOEDAS COM IA
# AUTOR: (Teu nome)
# DATA: Outubro 2025
# ============================================

# === 1. IMPORTAR BIBLIOTECAS ===
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# === 2. BAIXAR DADOS DO BITCOIN ===
print("ðŸ“¥ Baixando dados do Bitcoin (BTC-USD)...")
data = yf.download("BTC-USD", start="2023-01-01", end="2024-01-01", interval="1d")

# Visualizar
print(data.head())

# === 3. CRIAR FEATURES ===
data["Return"] = data["Close"].pct_change() * 100
data["MA5"] = data["Close"].rolling(window=5).mean()
data["MA10"] = data["Close"].rolling(window=10).mean()

# RSI simples (forÃ§a relativa)
delta = data["Close"].diff()
gain = (delta.where(delta > 0, 0)).rolling(14).mean()
loss = (-delta.where(delta < 0, 0)).rolling(14).mean()
rs = gain / loss
data["RSI"] = 100 - (100 / (1 + rs))

# === 4. CRIAR VARIÃVEL ALVO ===
# Se o preÃ§o de amanhÃ£ for maior â†’ 1 (subida), senÃ£o 0 (queda)
data["Target"] = np.where(data["Close"].shift(-1) > data["Close"], 1, 0)

# Remover NaN
data = data.dropna()

# === 5. SEPARAR FEATURES E ALVO ===
features = ["Return", "MA5", "MA10", "RSI"]
X = data[features]
y = data["Target"]

# Normalizar dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# === 6. DIVIDIR TREINO/TESTE ===
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, shuffle=False)

# === 7. TREINAR MODELO ===
model = DecisionTreeClassifier(max_depth=5, random_state=42)
model.fit(X_train, y_train)

# === 8. AVALIAR ===
y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f"ðŸ“Š AcurÃ¡cia do modelo: {acc*100:.2f}%")
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

# === 9. VISUALIZAR RESULTADOS ===
data_test = data.iloc[-len(y_test):].copy()
data_test["Pred"] = y_pred

plt.figure(figsize=(12,6))
plt.plot(data_test.index, data_test["Close"], label="PreÃ§o BTC")
plt.scatter(data_test.index, data_test["Close"], c=data_test["Pred"], cmap="bwr", label="PrediÃ§Ã£o (0=queda,1=subida)", alpha=0.7)
plt.title("ðŸ“ˆ PreÃ§o do Bitcoin com PrediÃ§Ãµes do Modelo")
plt.legend()
plt.show()

# === 10. CONCLUSÃƒO ===
print("âœ… O modelo conseguiu prever com precisÃ£o razoÃ¡vel as tendÃªncias de curto prazo do preÃ§o do Bitcoin.")
print("âš ï¸ Importante: o modelo Ã© apenas educativo e nÃ£o deve ser usado para decisÃµes financeiras reais.")
```
